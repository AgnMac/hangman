```python
sarasas=['miškas', 'lenta','vyras','karjera','pakvietimas','konteineris','moksleivis','mišrainė', 'lazanija','atostogos','šermukšnis','elnias','benzinas', 'glaistas', 'guma', 'lakas', 'linoleumas', 'konjakas','audinys', 'audeklas', 'išdaiga', 'kirtis', 'randas', 'miškas', 'javai', 'molis', 'duona', 'tešla', 'žemė', 'vynas', 'vėjas', 'velnias', 'brolis', 'arklys', 'vaisius', 'pavojus', 'žvėris', 'debesis','akmuo','šuo']
```

```python
import random
from IPython.display import clear_output
```

```python
#funkcija iš sąrašo išrenka betkurį žodį atsitiktinai
def zodis_spejimui():
    zodis=random.choice(sarasas)
    return zodis.upper() #atsitiktinį žodį grąžina didžiosiomis
```

```python
def zaidimas(zodis):
    zodzio_uzbaigimas = '*' * len(zodis) #tai eilute, kuria ismeta, kad zmogus matytu, ka speja. bruksniukai atitinka slepiamas raides
    guessed = False
    spetos_raides = []
    speti_zodziai = []
    bandymai = 6 #nusakys, kiek kuno daliu lieka iki pakarimo
    #sudaro: galva, kunas, 2rankos,2kojos
    print (f'Sužaiskime kartuves! Spėkite raidę arba žodį. Paslėptas žodis susideda iš {len(zodis)} raidžių')
    print(pakaruoklis(bandymai))
    print(zodzio_uzbaigimas)
    print('\n')
    #while loopas veikia kol bus atpetas zodis 
    # arba kol zaidejas isnaudoja visus spejimus
    while not guessed and bandymai > 0 :
        clear_output(wait=True)
        spejimas = input('Spėkite raidę arba žodį:').upper()
        #trys galimi variantai:raide, zodis arba iraso nesamone/nieko neiraso
        if len(spejimas) == 1 and spejimas.isalpha():
            #tikrinu ar raide jau buvo speta, ar yra zodyje ar nera
            if spejimas in spetos_raides:
                print(f"Raidę {spejimas} jau spėjote. Spėkit kitą raidę")
            elif spejimas not in zodis:
                print(f'Raidės {spejimas} nėra paslėptame žodyje.')
                bandymai-= 1
                spetos_raides.append(spejimas)
            else:
                print('Valio! Atspėjote raidę!')
                spetos_raides.append(spejimas)
                zodis_kaip_sarasas = list (zodzio_uzbaigimas)
                indeces = [i for i, raide in enumerate(zodis) if raide == spejimas ]
                for index in indeces:
                    zodis_kaip_sarasas[index] = spejimas
                zodzio_uzbaigimas = "".join(zodis_kaip_sarasas)
                if "*" not in zodzio_uzbaigimas:
                    guessed = True
        elif len(spejimas) == len(zodis) and spejimas.isalpha():
            if spejimas in speti_zodziai:
                print('Jūs jau spėjote žodį', spejimas)
            elif spejimas != zodis:
                print (spejimas, "nėra paslėptas žodis.")
                bandymai -= 1
                speti_zodziai.append(spejimas)
            else:
                guessed = True
                zodzio_uzbaigimas = zodis
        else:
            print('Įveskite raidę arba žodį. Jūsų ką tik įvestas simbolis/kombinacija netinka.')
        print(pakaruoklis(bandymai))
        print(zodzio_uzbaigimas)
        if len(spetos_raides) != 0:
            print(f'Jūsų spėtos raidės:{spetos_raides}')
        if len(speti_zodziai) != 0:
            print(f'Jūsų spėti žodžiai:{speti_zodziai}')
        print('\n')
    if guessed:
        print('Valio! Atspėjote žodį! Laimėjote žaidimą!')
    else: 
        print (f'Deja, išnaudojote visus bandymus. Pakaruoklis pakartas. Paslėptas žodis buvo {zodis}. Pabandykite dar kartą.')

        
#bandymų skaičius (iš 6 vis po vieną atimsim, kai bus neteisingai atsakoma) 
# bus prilyginamas kažkuriam iš pakaruoklio etapų
def pakaruoklis(bandymai):
    etapai = [
        """
        ___________
        |         |
        |         O/
        |        /|
        |        // 
        |
        |
        |
        
        """,
        """
        ___________
        |         |
        |         O/
        |        /|
        |        / 
        |
        |
        |
        
        """,
        """
        ___________
        |         |
        |         O/
        |        /|
        |        
        |
        |
        |
        
        """,
        """
        ___________
        |         |
        |         O
        |        /|
        |        
        |
        |
        |
        
        """,
        """
        ___________
        |         |
        |         O
        |         |
        |        
        |
        |
        |
        
        """,
        """
        ___________
        |         |
        |         O
        |        
        |
        |
        |
        |

        """,
        """
        ___________
        |         |
        |         
        |        
        |        
        |
        |
        |
        
        """
    ]
    return etapai[bandymai]

def kartuves():
    #zodis = zodis_spejimui()
    #zaidimas(zodis)
    while True:
        zodis = zodis_spejimui()
        zaidimas(zodis)
        clear_output(wait=True)
        pasirinkimas_zaisti = input('Ar norite vėl sužaisti kartuves? (TAIP/NE)').upper()
        if pasirinkimas_zaisti != 'TAIP':
            break







```
